```python
media_url = 'https://raw.githubusercontent.com/plotly/datasets/master/gapminderDataFiveYear.csv'
```

```python
medals_data = pd.read_csv(media_url)
```

###### 

```python
medals_data.head()
```

```python
import urllib3
http = urllib3.PoolManager()
r = http.request('GET', media_url)
r.status
response = r.data

#defocdificandolo
str_data = response.decode('utf-8')

#se divide el tring en un aaray de filas separandolos por intros
lines = str_data.split(",")

cols_name = lines[0].split(",")
n_cols = len(cols_name)


#se genera un diccionario donde estara la informacion

counter = 0
main_dict = {}
for col in cols_name:
    main_dict[col] = []
    
#procesamos fila a fila la informacion para ir rellenando el diccioanrio
for line in lines:
    if(counter > 0):
        values = line.strip().split(",") 
        for i in range(len(cols_name)):
            main_dict[cols_name[i]].append(values[i])
       
    counter += 1
    
print("El data set tiene %d filas y %d columnas"%(counter,n_cols))

medals_df = pd.DataFrame(main_dict)
medals_data.head()
```
